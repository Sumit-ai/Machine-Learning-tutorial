## plot using matplotlib library 

```python
import numpy as np # for mathematical library 
import matplotlib.pyplot as plt # for plotting the results 
import seaborn as sns # for beautiful plots 
sns.set()
y = np.random.randint(100, size=100) # take random values 
y1 = np.random.randint(100, size=100) 
y2 = np.random.randint(100, size=100)
x = np.arange(len(x))# define x axis 
```

```python
plt.figure(figsize = [30,10]) # define figure size here 
# 
plt.plot(x,y, '-o',label = "lebel 1")
plt.plot(x,y1, '-*',label = "lebel 2")
plt.plot(x,y2, '-',label = "lebel 3")
plt.legend(fontsize=20)
plt.xlabel("X axis", fontsize=20)
plt.ylabel("Y axis", fontsize=20)
# Now save the figure 
plt.savefig("plot.png")
```

## different types of plots 

```python
plt.figure(figsize = [30,5]) # define figure size here 
plt.scatter(x,y1)
plt.xlabel("X axis", fontsize=20)
plt.ylabel("Y axis", fontsize=20)
```

```python
plt.figure(figsize = [30,10]) # define figure size here 
data = {'a': np.arange(150),
        'c': np.random.randint(0, 100, 150),
        'd': np.random.randn(150)}
data['b'] = data['a'] + 10 * np.random.randn(150)
data['d'] = np.abs(data['d']) * 150

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a',fontsize=20)
plt.ylabel('entry b',fontsize=20)
plt.show()
```

```python
# line plot by using sns library 
plt.figure(figsize=(16,6))
sns.lineplot(x,y)
```

## distribution plot

```python
plt.figure(figsize = [10,10]) # define figure size here 
y = np.random.normal(100, size=1000)
sns.distplot(y);
```

```python
sns.distplot(y, bins=20, kde=False, rug=True)
```

```python
from scipy import stats
sns.distplot(y, kde=False, fit=stats.gamma);
```

```python
y2 = np.random.randint(100, size=1000)
x = np.random.randint(100, size=1000)# define x axis 
sns.jointplot(x,y2);
```

```python
plt.figure(figsize = [10,10])
sns.jointplot(x,y2, kind="kde");
plt.show()
```

```python
g = sns.jointplot(x,y2, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

```python
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(x,y2, cmap=cmap, n_levels=60, shade=True);
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(x, y2, ax=ax)
sns.rugplot(x, color="g", ax=ax)
sns.rugplot(y2, vertical=True, ax=ax);
```
